# Popularity Based Recommender System

```python
books.head(2)
```

```python
ratings.head(2)
```

```python
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.head(2)
```

```python
num_ratings_df = ratings_with_books.groupby('Book-Title', as_index=False)['Book-Rating'].count().rename(columns={'Book-Rating':'num_ratings'})
num_ratings_df
```

```python
avg_ratings_df = ratings_with_books.groupby('Book-Title', as_index=False)['Book-Rating'].mean().rename(columns={'Book-Rating':'avg_ratings'})
avg_ratings_df
```

```python
popular_df = num_ratings_df.merge(avg_ratings_df, on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

```python
popular_df.head()
```

# Collaborative Filtering Based Recommender System

We will not consider every book & every user because we want people who read alot & rate as well
1. we only consider those people who rate more than 200 books
2. the books would be those who got minimum 50 ratings

```python
ratings_with_books.head(3)
```

```python
x = ratings_with_books.groupby('User-ID').count()['Book-Title'] > 200
users_with_200_ratings = x[x].index
```

```python
users_with_200_ratings.nunique()
```

```python
filtered_ratings = ratings_with_books[ratings_with_books['User-ID'].isin(users_with_200_ratings)]
```

```python
filtered_ratings
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
famous_books.shape
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings.duplicated().sum()
```

```python
final_ratings.head()
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.shape
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('The Da Vinci Code')
```
